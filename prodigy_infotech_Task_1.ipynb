```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
countries = [
    "Afghanistan", "Albania", "Andorra", "Argentina",
    "Armenia", "Australia", "Austria", "Azerbaijan",
    "Burundi", "Belgium"
]
populations_2023 = [
    41454761, 2745972, 80856, 45538401,
    2990900, 26658948, 9131761, 10153958,
    13689450, 11787423
]

# Create bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=countries, y=populations_2023, palette="viridis")
plt.title("Population of Selected Countries (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```
